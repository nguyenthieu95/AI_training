```python
import GA_func as ga
import numpy as np
import matplotlib.pyplot as plt
import operator
import time
%matplotlib inline
from GA_func import *
```

```python
sample = GA(size_var = 50, size_pop = 210, parent_num = 21, tour_size = 20, constraints = [-10, 10],\
            mutation_rate = 0.0005)
```

```python
start = time.clock()
SM_value = sample.run(swapMutation)
SM_time = time.clock() - start
```

```python
min(SM_value)
```

```python
start = time.clock()
ScM_value = sample.run(scrambleMutation)
ScM_time = time.clock() - start
```

```python
min(ScM_value)
```

```python
start = time.clock()
Inv_value = sample.run(inverseMutation)
Inv_time = time.clock() - start
```

```python
min(Inv_value)
```

```python
start = time.clock()
Rr_value = sample.run(randomResetting)
Rr_time = time.clock() - start
```

```python
min(Rr_value)
```

```python
plt.figure(figsize = [20,10])
plt.title("Genetic Algorithms" )
plt.ylim(bottom = -25, top = 25)
plt.plot(SM_value, label = "SM - best_fit: %d (%.4lf s/epoch)" %(min(SM_value)*1000, SM_time/3000))
plt.plot(ScM_value, label = "BFM - best_fit: %d (%.4lf s/epoch)" %(min(ScM_value)*1000, ScM_time/3000))
plt.plot(Inv_value, label = "ScM - best_fit: %d (%.4lf s/epoch)" %(min(Inv_value)*1000, Inv_time/3000))
plt.plot(Rr_value, label = "IM - best_fit: %d (%.4lf s/epoch)" %(min(Rr_value)*1000, Rr_time/3000))
plt.grid(True)

leg = plt.legend(loc = 'upper right', ncol = 1,  shadow = True, fancybox = True)
leg.get_frame().set_alpha(0.5)

plt.xlabel("Generation")
plt.ylabel("Value")
```
