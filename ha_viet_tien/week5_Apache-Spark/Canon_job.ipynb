```python
sb_mean = df[df['SubmitTime'] != -1]['SubmitTime'].mean()
wt_mean = df[df['WaitTime'] != -1]['WaitTime'].mean()
rt_mean = df[df['RunTime'] != -1]['RunTime'].mean()
NPr_mean = df[df['NProc'] != -1]['NProc'].mean()
cpuT_mean = df[df['UsedCPUTime'] != -1]['UsedCPUTime'].mean()
um_mean = df[df['UsedMemory'] != -1]['UsedMemory'].mean()
stt_mean = df[df['Status'] != -1]['Status'].mean()
```

```python
# random for status = -1, keeping the mean
def random_status():
    if random.random() >= stt_mean:
        return 1
    return 0
```

```python
# convert -1 to something meaningful
def convert_minus_one(df):
    df.loc[df['SubmitTime'] == -1,'SubmitTime'] = sb_mean
    df.loc[df['WaitTime'] == -1,'WaitTime'] = wt_mean
    df.loc[df['RunTime'] == -1,'RunTime'] = rt_mean
    df.loc[df['NProc'] == -1,'NProc'] = NPr_mean
    df.loc[df['UsedCPUTime'] == -1,'UsedCPUTime'] = cpuT_mean
    df.loc[df['UsedMemory'] == -1,'UsedMemory'] = um_mean
    df.loc[df['Status'] == -1,'Status'] = random_status()
    
```

```python
# make time series data
def record_time_series(df, start, end, interval):
    prev = start
    cur = start + interval
    init_data = [[start/60, 0, 0]]
    columns = ['Time', 'CPU', 'Ram']
    time_df = pd.DataFrame(data = init_data, columns = columns)
    while(cur<=end):
        # take random number from normal sample
        x = random.uniform(prev, cur)
        #filter out
        mask = (df['StartTime'] <= x) & (df['EndTime'] >= x)
        new_df = df[mask]
        #caculate
        sum1 = new_df['NProc'].sum()
        sum2 = new_df['UsedMemory'].sum()
        
        """
        print("Prev: ", prev, "Cur: ", cur)
        print("Value Count:\n", new_df['NProc'].value_counts())
        """

        time_df = time_df.append({'Time':cur/60, 'CPU':sum1, 'Ram':sum2}, ignore_index=True)
        prev = cur
        cur = cur + interval
    return time_df
```

```python
interval = 50
```

```python
time_df = record_time_series(df, start, end, interval*60)
```
